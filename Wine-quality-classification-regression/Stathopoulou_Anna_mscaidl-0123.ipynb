```python
#Features
X = dataset.drop(['quality'], axis=1)

#Target (quality)
y = dataset['quality']
```

```python
from sklearn.model_selection import train_test_split

#Split the dataset in 80% train and 20% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)
```

# Regression
## Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt

#Evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R²:", r2)
```

```python
#Plot to compare the predicted quality to the true quality

plt.title("True quality / Predicted quality")
plt.xlabel("True quality")
plt.ylabel("Predicted quality")
plt.scatter(y_test, y_pred)
lims = [min(y_test.min(), y_pred.min()), max(y_test.max(), y_pred.max())]
plt.plot(lims, lims, 'r', lw=2)


plt.show()
```

# Regression
## Random Forest Regression


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
y_rf_pred = rf_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt

#Random Forest evaluation metrics
mae_rf = mean_absolute_error(y_test, y_rf_pred)
mse_rf = mean_squared_error(y_test, y_rf_pred)
rmse_rf = sqrt(mse_rf)
r2_rf = r2_score(y_test, y_rf_pred)

print("Random Forest MAE:", mae_rf)
print("Random Forest MSE:",mse_rf)
print("Random Forest RMSE:", rmse_rf)
print("Random Forest R²:", r2_rf)
```

```python
#Plot to compare the predicted quality to the true quality using Random Forest

plt.title("True quality / Predicted quality")
plt.xlabel("True Quality")
plt.ylabel("Predicted Quality")
plt.scatter(y_test, y_rf_pred, color="green")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "r", lw=2)


plt.show()

```

```python
### The Random Forest compared to Linear Regression achieved lower error values and higher R² showing more variance in wine quality
```

# Classification
## Logistic Regression

```python
import numpy as np

#Classification labels
##Wines with quality under 7 are labeled "0" (Not Good)
##Wines with quality over 7 are labeled "1" (Good)
dataset['label'] = np.where(dataset['quality'] >= 7, 1, 0)
```

```python
#Class distribution
##The dataset is imbalanced with 86% of "Not good" wines and 0.14% of "Good" wines
dataset['label'].value_counts(normalize=True)
```

```python
#Feaures label
X_clf = dataset.drop(['quality', 'label'], axis=1)

#Target label (0,1)
y_clf = dataset['label']
```

```python
from sklearn.model_selection import train_test_split

#Split the dataset in 80% train and 20% test
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42, stratify=y_clf
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score

#Logistic Regression Classifier
clf_model = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
clf_model.fit(X_train_clf, y_train_clf)

#Cross validation with F1 (5 fold cross validation). We use F1 because it balances precision and recall
scores = cross_val_score(clf_model, X_clf, y_clf, cv=5, scoring='f1')
print("F1 scores:", scores)
print("Mean F1 score:", scores.mean())

y_pred_log = clf_model.predict(X_test_clf)

#Confusion Matrix to show how many samples where classified correctly and how many were classified wrongly
ConfusionMatrixDisplay.from_estimator(clf_model, X_test_clf, y_test_clf)
plt.title("Confusion Matrix")
plt.show()

#Confusion Report to show F1 score accuracy, presicion
print(classification_report(y_test_clf, y_pred_log))
```

# Classification
## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#Random Forest Classifier with hyperparameters:
## n_estimators=500: number of trees
## max_depth=12: maximum depth of each tree
## min_samples_leaf=5 and
## min_samples_split=10 to control complexity of trees
## max_features='sqrt' random selection for splitting

rf_clf_model = RandomForestClassifier(
    n_estimators=500,
    max_depth=12,
    min_samples_leaf=5,
    min_samples_split=10,
    max_features='sqrt',
    class_weight={0:1, 1:4},
    random_state=42,
    n_jobs=-1
)
rf_clf_model.fit(X_train_clf, y_train_clf)

#Cross validation with F1 (5 fold cross validation). We use F1 because it balances precision and recall
scores = cross_val_score(rf_clf_model, X_clf, y_clf, cv=5, scoring='f1')
print("F1 scores:", scores)
print("Mean F1 score:", scores.mean())

y_pred_rf = rf_clf_model.predict(X_test_clf)

#Confusion Matrix to show how many samples where classified correctly and how many were classified wrongly
ConfusionMatrixDisplay.from_predictions(y_test_clf, y_pred_rf)
plt.title("Confusion Matrix")
plt.show()

#Confusion Report to show F1 score accuracy, presicion
print(classification_report(y_test_clf, y_pred_rf))
```

```python
# Overall the Random Forest Classifier performed better in classifying moth minority and majority classes.
```
